```python
# 1. import
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
# 2. agent blueprint
class AgentState(TypedDict):
    message: str
```

```python
# 3. the node
def compliment_node(state: AgentState) -> AgentState:
    state['message'] = state['message'] + ", your'e doing an amazing job learning LangGraph!"

    return state
```

```python
# 4 setting up the graph
graph = StateGraph(AgentState)
graph.add_node("compliment", compliment_node)
```

```python
# 5 connect the flow
graph.add_edge(START, "compliment")
graph.add_edge("compliment", END)

app = graph.compile()
```

```python
initail_memory = {"message": "Bob"}

result = app.invoke(initail_memory)
result["message"]
```
