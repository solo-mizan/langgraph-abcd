```python
# 1. import 
from typing import TypedDict, List
from langgraph.graph import StateGraph
```

```python
# 2. agent state schema
class AmarAgent(TypedDict):
    name: str
    values: List[int]
    operation: str
    answer: str
```

```python
# 3. node functions
def task_node(state: AmarAgent) -> AmarAgent:
    if "+" in state["operation"]:
        state["answer"] = f"Hi {state['name']}, your answer is: {sum(state['values'])}'"
    if "*" in state['operation']:
        res = 1
        for value in state['values']:
            res *= value
        state["answer"] = f"Hi {state['name']}, your answer is: {res}"
    return state
```

```python
# 4. graph
graph = StateGraph(AmarAgent)
graph.add_node('task', task_node)

graph.set_entry_point('task')
graph.set_finish_point('task')

app = graph.compile()
```

```python
# 5. invoke
result = app.invoke({'name': 'Jack Sparrow', 'values': [1, 2, 3, 4, 5], 'operation': "+"})
```
