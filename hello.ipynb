```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END # We imported START and END here

# 1. The Blueprint
class AgentState(TypedDict):
    message: str

# 2. The Node (Step)
def greeting_node(state: AgentState) -> AgentState:
    # Note: I added a space after "Hey " so the output looks a bit cleaner!
    state['message'] = "Hey " + state['message'] + ", how is your day going?"
    return state

# 3. Setting up the Graph
graph = StateGraph(AgentState)
graph.add_node("greeter", greeting_node)

# ==========================================
# --- THE NEW PIECES TO RUN THE AGENT ---
# ==========================================

# 4. Connect the Flow (Edges)
graph.add_edge(START, "greeter")
graph.add_edge("greeter", END)

# 5. Compile the Graph
app = graph.compile()

# 6. Run the Agent (Invoke)
initial_memory = {"message": "Tanvir"}
final_result = app.invoke(initial_memory)

# 7. See the Output
print(final_result)
```
